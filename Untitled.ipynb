# Bivariate Analysis

#### In this we are comparing sub categories of each category with the ratings, we can identify products having higher ratings.

```python
for i in df['category'].unique():
    sns.stripplot(x=df[df['category']==i]['sub_category'],y=df[df['category']==i]['rating'])
    plt.xticks(rotation='vertical')
    plt.show()    
```

## Observations:

1. these products have top ratings
2. these products have ratings not less than 3
3. these products have less sales etc

```python
df.head()
```

```python
plt.scatter(df['market_price'],df['rating'],s=df['rating']/1.5)
plt.xlabel('market_price')
plt.ylabel('rating_score')
plt.show()
```
